```python
def dot_product2(a,b):
    c = 0
    for i in range(len(a)):
        c += a[i]*b[i]
    return c


def phi(b):
    N = len(b)
    m = 0.2
    Ab = [0 for i in range(N)]
    n = int(math.sqrt(N))
    for i in range(N):
        x = i % n
        y = i // n
        x_plus = x + 1
        if x_plus > n-1:
            x_plus = 0
        x_minus = x - 1
        if x_minus < 0:
            x_minus = n-1
        y_plus = y + 1
        if y_plus > n - 1:
            y_plus = 0
        y_minus = y - 1
        if y_minus < 0:
            y_minus = n - 1
        Ab[i] = (b[n*y + x_plus] + b[n*y + x_minus] - 4*b[n*y+x] + b[n*y_plus+x] + b[n*y_minus+x])/2 + m**2 * b[n*y + x]
    return Ab


def CG_for_Sparce_Fn(N, tolerance = math.pow(10, -6)):
    B = []
    Residue = []
    for e in range(2):
        Residue += [[]]
        x = [1 for i in range(N)]    
        r = [0  for i in range(N)]
        d = [0  for i in range(N)]
        for i in range(N):
            if i==e: r[i] = 1 - phi(x)[i]
            else: r[i] = -phi(x)[i]
            d[i] = r[i]
        p = dot_product2(r,r)
        Residue[e] += [p**0.5]
        Trig = False
        k = 1
        while Trig == False:
            q = 0
            for l in range(N):
                q += d[l]*phi(d)[l]
            alpha = dot_product2(r,r)/q
            for j in range(N):
                x[j] += alpha*d[j]
                r[j] -= alpha*phi(d)[j]
            if dot_product2(r,r)**0.5 < tolerance:
                Trig = True
            else:
                beta = dot_product2(r,r)/p
                p = dot_product2(r,r)
                Residue[e] += [p**0.5]
                for j in range(N):
                    d[j] = r[j] + beta*d[j]
                k += 1
        B += [x]
    return B, Residue
    
N = 20**2
B, Res = CG_for_Sparce_Fn(N) 

iter = [[i+1 for i in range(len(Res[j]))] for j in range(len(Res))]


plt.plot(iter[0], Res[0], '--g.', label = 'Coloum_1')
plt.plot(iter[1], Res[1], '--r.', label = 'Coloum_2')
plt.title('Residue plot of Gauss Jacobi and Gauss Siedel')
plt.xlabel('Iteration')
plt.ylabel('Residue')
plt.legend()
plt.show()

Res_log = []
for i in range(len(Res[0])):
    Res_log += [math.log10(Res[0][i])]

plt.plot(iter[0], Res_log, '--g.', label = 'Coloum_1')
plt.title('Logerithm of Residue plot of Gauss Jacobi and Gauss Siedel')
plt.xlabel('Iteration')
plt.ylabel('log of Residue')
plt.legend()
plt.show()


#### Printing 1st Two col of inverse matrix
for i in range(N):
    print(B[0][i], '  |  ', B[1][i])
```
